```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
coeffs = pd.read_csv(r'coefficients.csv', header=None)
```

```python
plt.figure(figsize=(9,6))
plt.plot(coeffs.iloc[15:,:], marker='o', alpha=.5, color='purple')
plt.xlabel('Stage')
plt.ylabel('Coefficient')
plt.title("Clustering coefficient plot", fontsize=21)
plt.show()
```

```python
t_a = [-.46030, -.48428, 1.52877]
bjc = [-1.09448,.47632,.45394]
m_a = [-.99766,.42427,.41934]
```

```python
clstr_means = pd.DataFrame({'BJC': bjc, 'TA': t_a, 'MA': m_a})
x = [1,2,3]
my_xticks = [1, 2, 3]
```

```python
clstr_means
```

```python
plt.plot(x, clstr_means.BJC, marker='o', color='darkblue', alpha=.5, label='Belief in Jewish Consp.')
plt.plot(x, clstr_means.TA, marker='o', color='teal', alpha=.5, label='Traditional A-S')
plt.plot(x, clstr_means.MA, marker='o', color='orange', alpha=.5, label='Modern A-S')
plt.xticks(x, my_xticks)
plt.xlabel('Cluster number')
plt.ylabel('Mean')
plt.legend()
plt.show()
```
